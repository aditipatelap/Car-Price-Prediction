```python
# Plot the box plot of the 'price' column to find outliers
plt.figure(figsize=[3,4])
df1.boxplot('price')
plt.title("Car Price Distribution Box-Plot")
```

```python
# Find how many outlier are there in (numeric)column - using z score normalization
def findOutliers(df):
    # Select numeric columns
    numeric_columns = df.select_dtypes(include=['number']).columns
    print("selected columns are:\n", numeric_columns)
    
    # Set the threshold for Z-scores
    threshold = 2
    
    for column in numeric_columns:
        z_scores = (df[column] - df[column].mean()) / df[column].std()
        outliers = df[z_scores.abs() > threshold]
        print("Outliers in column '{}':".format(column), len(outliers))
        
findOutliers(df1)
```

## Training & Testing

Let's first split the dataset into train and test datasets.

- **from sklearn.metrics import r2_score**: It's use for measuring **regression** score.
- **from sklearn.metrics import accuracy_score**: It's use for measuring **classification** score.

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import train_test_split

#slicing the dataset and reomoving the selling price for training the model
x = df2.iloc[:, :-1]
#storing the selling price for checking..as this is the value to be predicted
y = df2.iloc[:, -1]

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state = 50)
```

### Linear Regression model

Linear regression is a statistical model that predicts a dependent variable from one or more independent variables.\
The equation for simple linear regression is:
y = b0 + b1 * x

```python
from sklearn.linear_model import LinearRegression

linearRegression = LinearRegression()
linearRegression.fit(xtrain, ytrain)
predictions = linearRegression.predict(xtest)
accuracy_linear = r2_score(ytest, predictions)
```

### Decision Tree Regression model

Decision tree regression is a supervised learning algorithm that uses a tree-like structure to predict a continuous value. The tree is built by recursively splitting the data into smaller and smaller subsets, until each subset contains only data points that are very similar to each other. The prediction for a new data point is then made by following the tree from the root node to the leaf node that corresponds to the new data point.

```python
from sklearn.tree import DecisionTreeRegressor

decisionTreeRegressor = DecisionTreeRegressor()
decisionTreeRegressor.fit(xtrain, ytrain)
predictions = decisionTreeRegressor.predict(xtest)
accuracy_dtr = r2_score(ytest, predictions)
```

### Random Forest model

A random forest model is an ensemble learning method for classification, regression, and other tasks that operates by constructing a multitude of decision trees at training time. For classification tasks, the output of the random forest is the class selected by most trees. For regression tasks, the mean or average prediction of the individual trees is returned. The idea behind this ensemble approach is to reduce overfitting and improve the overall predictive performance. Random forests generally outperform decision trees, but their accuracy is lower than gradient boosted trees.

Here, are the steps on how random forest works:
1. Bootstrapping: 
Random samples are taken from the training dataset with replacement. This means that some data points may be selected more than once, while others may not be selected at all.
2. Building decision trees: 
Decision trees are built on each bootstrapped sample. Each tree is built using a random subset of features.
3. Voting: 
The predictions from each decision tree are combined using majority voting for classification tasks and averaging for regression tasks.

```python
from sklearn.ensemble import RandomForestRegressor

randomForestRegressor = RandomForestRegressor()
randomForestRegressor.fit(xtrain, ytrain)
predictions = randomForestRegressor.predict(xtest)
accuracy_rf = r2_score(ytest, predictions)
```

### Accuracy of the models:

```python
print("Linear Regression model:", "{:.4f}".format(accuracy_linear))
print("Random Forest model:", "{:.4f}".format(accuracy_rf))
print("Decision Tree Regression model:", "{:.4f}".format(accuracy_dtr))
```

Here we can see Random Forest model gives the higher accuracy of the testing dataset.\
Let's select the model and save it.

## Save Model

```python
from joblib import dump

# Save the model to a file
filename = 'model.joblib'
dump(randomForestRegressor, filename)
```
